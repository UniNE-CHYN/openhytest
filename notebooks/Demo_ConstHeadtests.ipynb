First, we append the openhytest folder to the system path using the 'sys' library and import than the 'openhytest' library as ht. In addition, we import the 'matplotlib.pyplot' library for plotting and the 'pandas' library for data management.

```python
%matplotlib inline
import sys
sys.path.append("..")
import openhytest as ht
import matplotlib.pyplot as plt
import pandas as pd
```

 ## Load the data
 The data set for this example has been typed from: 
 

```python
data = pd.read_csv('../data/jlq_ds1.txt', header=None,sep=" ",names=['t','q'], dtype={'t': 'float64', 'q': 'float64'})
plt.plot(data.t,data.q,'o')
plt.xlabel('Time in seconds')
plt.ylabel('Flow rate m^3/s')
plt.show()
```

## Jacob-Lohman (1952) model

1) Load the Jacob-Lohman model with hydraulic head of 28.142 m and wellbore radius of 8.4 cm

2) Typecurve plot

```python
jcl = ht.JacobLohman(s=28.142, r=0.084, df = data)
jcl.plot_typecurve()
```

```python
jcl.guess_params()
```

```python
jcl.rpt(fitmethod='trf')
```
